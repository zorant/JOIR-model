### Create task class

```python
# Judgment of Imminence
class joi:

    def __init__(self, Nt=30000, i_dt=0.001, i_k=8, i_buff_len=100, i_g=1, i_tstr_min=0.05, i_tstr_max=2, e_thr = 0.01):
        
        """
        Initialize JOI task
        
        Keyword arguments:
        Nt -- number of time points 
        i_dt -- time step (for input nodes)
        i_k -- value of k (for input nodes)
        i_buff_len -- number of taustar nodes (for input nodes)
        i_g -- parameter g (determinizes scaling of tilde{f} - g=1->equal amplitude, g=0->power-law decay of amplitude)
        i_tstr_min -- smallest taustar (for input nodes)
        i_tstr_max -- largest taustar (for input nodes)
        
        e_thr -- evidence accumulator threshols
        """
        
        self.Nt = Nt 
        
        # Input parameters
        self.i_dt = i_dt
        self.i_k = i_k
        self.i_buff_len = i_buff_len
        self.i_g = i_g
        self.i_tstr_min = i_tstr_min
        self.i_tstr_max = i_tstr_max
        
        
        # JOI specific parameters
        self.probe_size = 2 # how many items does the probe stimulus contain
        self.acc_type = 2 # two independent accumulators   
        
        self.e_thr = e_thr
        
        self.lag = []
        self.alpha_array = np.zeros(self.Nt)
        self.lags = []
        self.reaction_times = []
        self.answers = []
        self.reaction_time = 0.
        
```

### Update memory at every time step

```python
def update_input_memory(p_n, time_index):
    for name in p_n.input_names:
        p_n.inputs[p_n.input_names.index(name)].support.update(
            time_index, np.max([0, p_n.inputs[p_n.input_names.index(name)].value[time_index] - 
        p_n.inputs[p_n.input_names.index(name)].value[time_index-1]]))

def update_associative_memory(p_n, time_index):
    f = np.zeros((len(p_n.input_names), 1))
    til_f = np.zeros((len(p_n.input_names), p_n.i_buff_len))
    for input_no, name in enumerate(p_n.input_names):
        f[input_no] = p_n.inputs[p_n.input_names.index(name)].value[time_index]
        til_f[input_no,:] = p_n.inputs[p_n.input_names.index(name)].support.til_f[p_n.i_k:-p_n.i_k, time_index]
    p_n.A.update(f, til_f)
    
```

### Generate radom data

```python
def generate_joi_data(p_n):
    # Put a random letter
    sample_no = random.randrange(0, len(p_n.input_names))
    p_n.sample = p_n.input_names[sample_no]
    # Add two probes at the end
    input_names_tmp = ['R', 'Y', 'T', 'X', 'Q', 'G', 'K', 'Z', 'F', 'L', 'P', 
                       'N', 'R', 'Y', 'T', 'X', 'Q', 'G', 'K', 'Z', 'F', 'L', 'P', 'N']
    p_n.probes = []
    # firs probe is letter 1 to 5 steps in the future from the input letter (as in Singh et al)
    i = random.randrange(sample_no+1,sample_no+6)
    # second probe is letter at least 2 letters after the first probe and not more than 7 letters after
    # the input letter (similar to Singh et al)
    j = random.randrange(i+2,sample_no+8)
    
    p_n.probes.append(input_names_tmp[j])  # larger lag placed first 
    p_n.probes.append(input_names_tmp[i])
    p_n.lag_future = []
    p_n.lag_future.append(j-sample_no)
    p_n.lag_future.append(i-sample_no)
        
    #print(p_n.input_names[sample_no])
    #print(p_n.probes)
    #print(p_n.lag_future)
    
```

```python
def generate_joi_training_data(p_n):
    p_n.input_names = ['R', 'Y', 'T', 'X', 'Q', 'G', 'K', 'Z', 'F', 'L', 'P', 'N']
    p_n.inputs = []
    for name in p_n.input_names:
        _ = p_n.inputs.append(Node(name, tstr_min=p_n.i_tstr_min, tstr_max=p_n.i_tstr_max, 
                                            buff_len=p_n.i_buff_len, k=p_n.i_k, 
                                            dt=p_n.i_dt, g=p_n.i_g, Nt=p_n.Nt))
    # Zero all inputs
    for name in p_n.input_names:
        p_n.inputs[p_n.input_names.index(name)].value = np.zeros(p_n.Nt)
    # Make a long sequence with transition probabilities same as in Singh et al
    i = 0
    while i in range(288):
        p_n.inputs[np.mod(i,12)].value[(i+1)*100] = 1
        if np.random.uniform(0,1,1)>0: #0.1
            i = i + 1
        else:
            i = i + np.int(np.round(np.random.uniform(2,12,1)))
```

### Run the simulation

```python
p_n = joi() #task name
generate_joi_training_data(p_n) # generate training data
p_n.A = Associate(p_n.i_buff_len, len(p_n.input_names))
for time_index in range(p_n.Nt):
    update_input_memory(p_n, time_index)
    update_associative_memory(p_n, time_index)
```

```python
p_n.lags = []
p_n.reaction_times = []
p_n.answers = []
p_n.e_thr = 0.1 #50

ntrials = 5000


for trial in range(ntrials): 
    noise_factor_l = random.uniform(0.15, 1.85)
    noise_factor_r = random.uniform(0.15, 1.85)
    #if trial % 10 == 0:
        #print(trial)
        #print('dumping')
        #dump(p_n, filename='p_n_JOI_tmp.gz')
    generate_joi_data(p_n) # generates the data for this trial
    # Initialize
    ev_l = 0.
    ev_r = 0.
    p_n.ev_left = 0
    p_n.ev_right = 0
    # Scan the memory slot by slot
    for g in range(p_n.i_buff_len):
        for name in p_n.input_names:
            if name == p_n.probes[0]:
                ev_l_tmp = p_n.A.M[g, p_n.input_names.index(name), p_n.input_names.index(p_n.sample)]
            if name == p_n.probes[1]:
                ev_r_tmp = p_n.A.M[g, p_n.input_names.index(name), p_n.input_names.index(p_n.sample)]
        ev_l = ev_l + ev_l_tmp*noise_factor_l
        ev_r = ev_r + ev_r_tmp*noise_factor_r
        if ev_l > p_n.e_thr or ev_r > p_n.e_thr:
            if ev_l > ev_r:
                p_n.ev_left = 1
            else:
                p_n.ev_right = 1
            break
    #print(ev_l)
    #print(ev_r)
    p_n.reaction_time = g # If not enough evidence before time limit
    # Append the results
    p_n.lags.append(p_n.lag_future)
    p_n.reaction_times.append(p_n.reaction_time)
    if p_n.ev_right:
        p_n.answers.append(0) # incorrect
    elif p_n.ev_left:
        p_n.answers.append(1) # corect
    else:
        p_n.answers.append(2) # time out - no answer 
```

```python
def make_joi_figures(p_n):
    
    # covert things into np arrays
    rt = np.array(p_n.reaction_times)
    lags = np.array(p_n.lags)
    answers = np.array(p_n.answers)
    # lag 0 less recent
    # lag 1 more recent
    # answer 0 correct
    # answer 1 incorect
    
    
    # lag vs RT - correct
    fig = pl.figure(figsize=(4, 3), dpi= 80, facecolor='w', edgecolor='k')
    ax = pl.subplot(111)
    rt_mean = np.zeros((7,7))
    rt_std = np.zeros((7,7))
    for j in range(7): # across more recent
        for i in range(7): # across less recent
            rt_mean[i,j] = np.mean(rt[np.where((lags[:,0]==i+1) & (lags[:,1]==j+1) & (answers==0))])
            rt_std[i,j] = np.std(rt[np.where((lags[:,0]==i+1) & (lags[:,1]==j+1) & (answers==0))])
        ax.errorbar(np.arange(1, 8), rt_mean[:,j], rt_std[:,j], marker='.', color=str(j/6.))
        #ax.plot(np.arange(1, 8), rt_mean[:,j], marker='.', color=str(j/6.))
    pl.ylabel('Response Time [au]', fontsize=14)
    pl.xlabel('Lag to the less imminent probe', fontsize=14) #Lag of less recent item
    #pl.title('Correct responses')
    #ax.set_yticklabels([])
    pl.savefig("JOI_less_recent_lag_vs_RT_noise2.pdf", bbox_inches='tight')
    pl.show()
    
    
    # lag vs RT - incorrect
    fig = pl.figure(figsize=(4, 3), dpi= 80, facecolor='w', edgecolor='k')
    ax = pl.subplot(111)
    rt_mean = np.zeros((7,7))
    rt_std = np.zeros((7,7))
    for j in range(7): # across more recent
        for i in range(7): # across less recent
            rt_mean[i,j] = np.mean(rt[np.where((lags[:,0]==i+1) & (lags[:,1]==j+1) & (answers==1))])
            rt_std[i,j] = np.std(rt[np.where((lags[:,0]==i+1) & (lags[:,1]==j+1) & (answers==1))])
        ax.errorbar(np.arange(1, 8), rt_mean[:,j], rt_std[:,j], marker='.', color=str(j/6.))
    pl.ylabel('Response Time [au]', fontsize=14)
    pl.xlabel('Lag to the less imminent probe', fontsize=14) #Lag of less recent item
    #pl.title('Incorrect responses')
    pl.savefig("JOI_inc_lag_vs_RT_noise2.pdf", bbox_inches='tight')
    pl.show()
    
    
    
    # Less recent lag vs accuracy
    fig = pl.figure(figsize=(4, 3), dpi= 80, facecolor='w', edgecolor='k')
    ax = pl.subplot(111)
    acc_mean = np.empty((7,7))
    acc_mean[:] = np.nan
    for j in range(7): # across more recent
        for i in range(7): # across less recent
            correct = np.sum((lags[:,0]==i+1) & (lags[:,1]==j+1) & (answers==0))
            total = np.sum((lags[:,0]==i+1) & (lags[:,1]==j+1))
            if total > 0:
                acc_mean[i,j] = correct/total
        ax.plot(np.arange(1, 8), acc_mean[:,j], marker='.', color=str(j/6.))
    pl.ylabel('Accuracy', fontsize=14)
    pl.xlabel('Lag to the less imminent probe', fontsize=14) #Lag of less recent item
    pl.savefig("JOI_acc_lag_vs_RT_noise2.pdf", bbox_inches='tight')
    pl.show()
    
    
    
    # lag vs more recent RT - correct only
    rt_mean = np.zeros((7))
    rt_std = np.zeros((7))
    for i in range(7):
        rt_mean[i] = np.mean(rt[np.where((lags[:,1]==i+1) & (answers==0))])
        rt_std[i] = np.std(rt[np.where((lags[:,1]==i+1) & (answers==0))])
    fig = pl.figure(figsize=(4, 3), dpi= 80, facecolor='w', edgecolor='k')
    ax = pl.subplot(111)
    ax.plot(np.arange(1, 8), rt_mean, '.', color='k')
    ax.errorbar(np.arange(1, 8), rt_mean, rt_std, marker='.', fmt='none', ecolor='k')
    pl.ylabel('Response Time (au)', fontsize=14)
    pl.xlabel('Lag to the more imminent probe', fontsize=14) #Lag of more recent item
    log_lags_correct = np.log(np.squeeze(lags[np.where((answers==0)),1]))
    rt_correct = rt[np.where((answers==0))]
    slope, intercept, r_value, p_value, std_err = stats.linregress(log_lags_correct, rt_correct)
    lag_times = np.arange(1,7)
    ax.plot(lag_times, slope*np.log(lag_times) + intercept, color='k')
    ax.set_xscale("log", nonposx='clip')
    ax.set_xticks(np.arange(1,6)) 
    ax.set_xticklabels([str(i) for i in range(1,6)])
    #ax.set_ylim(-500, 1800)
    #ax.set_yticklabels([])
    pl.gca().xaxis.set_minor_formatter(matplotlib.ticker.NullFormatter())
    pl.savefig("JOI_more_recent_lag_vs_RT_noise2.pdf", bbox_inches='tight')
    pl.savefig("JOI_more_recent_lag_vs_RT_noise2.png", bbox_inches='tight')
    pl.show()
    
    # cummulative plots
    for i in range(7):
        #a = np.array(p_n.reaction_times)[np.where(np.array(p_n.lags)==i+2)[0]]
        a = rt[np.where((lags[:,1]==i+1) & (answers==0))]
        #print(a)
        N = len(a)
        #if N>0:
        #    a = np.random.normal(loc=np.mean(a), scale=p_n.s_sigma_ad+p_n.e_thr*np.mean(a), size=len(a))
        X2 = np.sort(a)
        F2 = np.array(range(N))/float(N)
        pl.plot(X2, F2)
    #ax.tick_params(labelsize=14)
    #pl.xlabel('Response time (arbitrary units)', fontsize=14)
    #pl.ylabel('Cumulative probability', fontsize=14)
```
